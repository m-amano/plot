```python
from pytrends.request import TrendReq
import pandas as pd
```

```python
pytrends = TrendReq(hl = 'ja-JP', tz=360)
kw_list1=["Python", "JavaScript", "Java"]
kw_list2=["Ruby", "PHP", "SQL"]
pytrends.build_payload(kw_list1, timeframe='2014-01-01 2019-09-22', geo ='JP')
df1 = pytrends.interest_over_time()
pytrends.build_payload(kw_list2, timeframe='2014-01-01 2019-09-22', geo ='JP')
df2 = pytrends.interest_over_time()
```

```python
df = pd.concat([df1, df2], axis=1)
df.plot(figsize=(20, 5), lw=.7)
```
